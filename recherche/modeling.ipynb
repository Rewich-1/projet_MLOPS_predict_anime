```python
import pandas as pd
import numpy as np  
import math
```

```python
df=pd.read_csv('../data/Anime_data.csv')
df
```

```python
animes=df[['Title','Genre','Synopsis','Type','Producer','Studio','Rating','ScoredBy']]
animes
```

```python
animes = animes.dropna(subset=['Genre','Synopsis','Producer','Studio','Rating'])
animes
```

```python
animes['Genre'] = animes['Genre'].apply(lambda x: x[1:-1].replace("'",''))
animes['Producer'] = animes['Producer'].apply(lambda x: x[1:-1].replace("'",''))
animes['Studio'] = animes['Studio'].apply(lambda x: x[1:-1].replace("'",''))
animes
```

```python
genre_encoded = animes['Genre'].str.get_dummies(sep=',')
producer_encoded = animes['Producer'].str.get_dummies(sep=',')  
studio_encoded = animes['Studio'].str.get_dummies(sep=',')
type_encoded = animes['Type'].str.get_dummies()
```

```python
def normalize_genres(row):
    sum_row = row.sum() if row.sum() != 0 else 1
    return [x/math.sqrt(sum_row) for x in row]

animes['Genre']=genre_encoded.apply(normalize_genres, axis=1)  
animes['Producer']=producer_encoded.apply(normalize_genres, axis=1)  
animes['Studio']=studio_encoded.apply(normalize_genres, axis=1)  
animes['Type']=type_encoded.apply(normalize_genres, axis=1)  
animes
```

```python
from sentence_transformers import SentenceTransformer

sbert = SentenceTransformer ('distilbert-base-nli-mean-tokens')
```

```python
t_emb = sbert.encode(animes['Title'].values , show_progress_bar =True)
s_emb = sbert.encode(animes['Synopsis'].values , show_progress_bar =True)
```

```python
animes['Title'] = t_emb.tolist()    
animes['Synopsis'] = s_emb.tolist()    
animes
```
