```python
import pandas as  pd
import numpy as  np
```

```python
df = pd.read_csv('../data/Anime_data.csv')
```

```python
df
```

```python
df = df["Studio"]
```

```python
df
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
import ast

list_studio = []
for x in df:
    x = ast.literal_eval(x)
    for j in x:
        if j not in list_studio:
            list_studio.append(j)
```

```python
list_studio
```

```python
import json


with open("../data/list_studio", "w") as fp:
    json.dump(list_studio, fp)
```
