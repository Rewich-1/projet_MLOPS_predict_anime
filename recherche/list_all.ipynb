```python

import pandas as pd 
import re
```

```python
   
df=pd.read_csv('../data/Anime_data.csv')
df
```

```python
animes=df[['Title','Genre','Synopsis','Type','Producer','Studio','Rating','ScoredBy']]
animes
```

```python
animes = animes.dropna(subset=['Genre','Rating','Synopsis'])
animes
```

```python
animes['Producer'] = animes['Producer'].fillna("[Unknown]")
animes['Studio'] = animes['Studio'].fillna("[Unknown]")
animes
```

```python
animes['Genre'] = animes['Genre'].apply(lambda x: x[1:-1].replace("'",''))
animes['Producer'] = animes['Producer'].apply(lambda x: x[1:-1].replace("'",''))
animes['Studio'] = animes['Studio'].apply(lambda x: x[1:-1].replace("'",''))


animes['Genre'] = animes['Genre'].apply(lambda x: re.sub(r'\s*,\s*', ',', x))
animes['Producer'] = animes['Producer'].apply(lambda x: re.sub(r'\s*,\s*', ',', x))
animes['Studio'] = animes['Studio'].apply(lambda x: re.sub(r'\s*,\s*', ',', x))
animes
```

```python
list_genre  = (", ".join(animes['Genre'])).split(', ')
list_genre = list(set(list_genre))
len(list_genre)
```
